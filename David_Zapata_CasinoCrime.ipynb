```python
import requests
import json

gkey = 
target_city = "Los Angeles County, CA"
target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
    'address={0}&key={1}').format(target_city, gkey)
geo_data = requests.get(target_url).json()
#print(json.dumps(geo_data, indent=4, sort_keys=True))
print(geo_data["results"][0]["geometry"]["location"])
```

```python
params = {"key": gkey}

Latdict = {}
Lngdict = {}

for target_city in unique:
    
    target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
        'address={0}&key={1}').format(target_city, gkey)
    geo_data = requests.get(target_url).json()
    #print(json.dumps(geo_data, indent=4, sort_keys=True))
    print(geo_data["results"][0]["geometry"]["location"])
    Latdict[target_city]=geo_data["results"][0]["geometry"]["location"]["lat"]
    Lngdict[target_city]=geo_data["results"][0]["geometry"]["location"]["lng"]
```

```python
csvpath_df["Lat"] = csvpath_df["Countym"].map(Latdict)
csvpath_df["Lng"] = csvpath_df["Countym"].map(Lngdict)
csvpath_df
#filterout N/a
```

```python
import matplotlib.pyplot as plt
import gmaps
```

```python
locations = csvpath_df.loc[(csvpath_df["Lat"].notnull())&(csvpath_df["Lng"].notnull())][["Lat","Lng"]]
#rating = csvpath_df['Count'].astype(int)
#docloc
#condition 
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, 
                                 dissipating=False, max_intensity=.75,
                                 point_radius=.8)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
#please note that if you do run this cell; please be aware that it could be taxing in time and computer resources.
```
