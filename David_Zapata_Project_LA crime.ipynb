```python
# Dependencies
import os
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
import requests
import json
import gmaps

# Google API Key
from config import gkey
```

```python
csvpathLA = os.path.join('Crime_Data_from_2010_to_2019.csv')
```

```python
csvpathLA_df = pd.read_csv(csvpathLA)
csvpathLA_df.head()
```

```python
csvpathLA_df.columns
```

```python
locations = csvpathLA_df.loc[(csvpathLA_df["LAT"].notnull())&(csvpathLA_df["LON"].notnull())][["LAT","LON"]]
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, 
                                 dissipating=False, max_intensity=.05,
                                 point_radius=.08)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig    
#please note that run this cell had taken me Twenty (20) minutes to render. I will leave it blank for that issue.
```
